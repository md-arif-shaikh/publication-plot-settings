```python
import plotsettings as plts
import numpy as np
```

```python
x = np.arange(-np.pi, np.pi, 0.01)
```

```python
fig, ax = plts.set("APS")
for idx in range(5):
    ax.plot(x, (x + idx), label="$%.f$" % idx)
ax.legend(loc="upper left")
ax.grid()
```

```python
import plotsettings as plts
fig, ax = plts.set("Notebook")
for idx in range(5):
    ax.plot(x, (x + idx), label="$%.f$" % idx)
ax.legend(loc="upper left")
ax.grid()
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$x + i$")
```
