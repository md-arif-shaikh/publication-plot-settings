```python
import plotsettings as plts
import numpy as np
```

```python
x = np.arange(-np.pi, np.pi, 0.01)
```

```python
fig, ax = plts.set("PRD")
for idx in range(10):
    ax.plot(x, np.sin(x + idx))
ax.grid()
```

```python
fig, ax = plts.set("APJ")
for idx in range(10):
    ax.plot(x, np.sin(x + idx))
ax.grid()
```

```python
fig, ax = plts.set("Presentation")
for idx in range(10):
    ax.plot(x, np.sin(x + idx))
ax.grid()
```

```python
fig, ax = plts.set("Notebook")
for idx in range(10):
    ax.plot(x, np.sin(x + idx))
ax.grid()
ax.set_xlabel(r"Angle [$\theta$]")
ax.set_ylabel(r"$\sin (x)$")
```

```python
fig, ax = plts.set("PRD", fig_type="twocol")
for idx in range(10):
    ax.plot(x, np.sin(x + idx))
ax.grid()
```
